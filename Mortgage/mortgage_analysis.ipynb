```python
import pandas as pd
import requests

# series code for 2 year fixed 95% LTV
series_code = "IUM2WTL"

url = f"https://api.bankofengland.co.uk/boeapps/historic-interest-rates-and-macro-data/series/{series_code}"

response = requests.get(url)

data = response.json()

df = pd.DataFrame(data["observations"])
df.head()
```
